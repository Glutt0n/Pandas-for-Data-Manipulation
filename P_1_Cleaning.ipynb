# DATA CLEANING

```python
# importing packages
import pandas as pd
```

```python
# Reading in and displaying data
df=pd.read_csv('/Users/oliver/Desktop/datasets/sales_predictions_2023.csv')
display(df.head(),df.info())
```

```python
## Identifying missing values
# displaying null values
display(df.loc[df.isna().any(axis=1)]) 

# getting percentage per columns
print((df.isna().sum()/len(df))*100) 
# getting total null count
df.isna().sum().sum() 
```

    It appears that the data has many missing values. The Item_Weight and Outlet_Size column's missing values compose ~17% and ~28% of the data respectively. These values will be imputed with a placeholder value 'MISSING' representing missing data. 

```python
# Imputing Outlet_Size with "MISSING"
df['Outlet_Size'].fillna(value='MISSING',inplace=True)

# Imputing Item_Weight with mean weight
df['Item_Weight'].fillna(value=df.Item_Weight.mean(),inplace=True)

# Confirming there are no missing values
df.isna().sum().sum()
```

```python
# Looking for nonsensical values in numeric columns
df.describe()
```

All numeric values look feasible.

```python
# Looking for any inconsistent values in categorical data
for col in df.select_dtypes('object').columns:
    print(col,':\n',df[col].unique(),'\n')
```

Some things that we see from this loop is that:
1. Item_Identifier appears to be filled with many, likely unuseful, unique values.

2. Item_Fat_Content contains inconsistent values.

```python
# displaying Item_Identifier
df['Item_Identifier'].nunique()
```

The excessive amount of unique values is useless for machine learning. This feature will be dropped.

```python
# Dropping column
df=df.drop(columns='Item_Identifier')
# Confirming drop
print(df.columns=='Item_Identifier')
```

Checking for duplicates

```python
# Checking for duplicates
df.loc[df.duplicated()]
# getting total
df.duplicated().sum() 
```

There are no duplicates to drop.
